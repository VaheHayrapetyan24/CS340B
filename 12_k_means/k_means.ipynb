```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from sklearn.metrics import adjusted_rand_score
```

```python
X, Y = make_blobs(n_samples=3000, centers=10)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=50)
```

```python
cost = []
r = range(1,20)
for num_clusters in list(r):
    model = KMeans(n_clusters=num_clusters, max_iter=10000)
    model.fit(X)
    cost.append(model.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(r, cost, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
model = KMeans(n_clusters=10, max_iter=10000)
model.fit(X)
y_predict = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_predict, s=50)
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()

print(adjusted_rand_score(Y, y_predict))
```
