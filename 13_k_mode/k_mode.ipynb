```python
import numpy as np
import pandas as pd
from kmodes.kmodes import KModes # I didn't find kmodes in sklearn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./soybean_large/soybean-large.data', header=None) 

data = data.iloc[:, :-1]
```

```python
cost = []
for num_clusters in list(range(1, 20)):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 1, verbose=0)
    kmode.fit_predict(data)
    cost.append(kmode.cost_)

plt.plot(range(1,20), cost, marker='o')
plt.show()
```

```python
optimal_k = 3

kmode = KModes(n_clusters=optimal_k, init = "random", n_init = 1, verbose=0)
clusters = kmode.fit_predict(data)

print(clusters)
print(kmode.cluster_centroids_)
```
