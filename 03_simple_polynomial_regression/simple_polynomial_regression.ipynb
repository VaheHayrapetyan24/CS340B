```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y = np.array([2, 3.5, 6, 8, 7, 8.2, 10, 7.9, 6, 5, 4.5, 3])
X = np.array([1, 1.5, 2, 2.5, 3, 3.3, 4, 4.2, 5, 4.8, 5.2, 4.9])

P_degree = 7
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
powers = np.arange(P_degree)
X_powers = X[:, None]**powers
print(X_powers)

def linear_regression(l, iter, X, Y):
    theta = np.ones(P_degree)
    for j in range(iter):
        # print((Y - X_powers.dot(theta)).dot(X_powers))
        # print((Y - X_powers.dot(theta)).dot(X_powers)*l/len(Y))
        theta = theta + (Y - X_powers.dot(theta)).dot(X_powers)*l/len(Y)
    return theta
```

```python
theta = linear_regression(0.0000000001,1000000000,X,Y)
Y_predict = X_powers.dot(theta)
print(Y_predict)
print(theta)
```

```python
plt.scatter(X, Y)
plt.scatter(X, Y_predict)
plt.show()
```

```python
mean = np.sum(Y) / len(Y)
SST = np.sum((Y - mean)**2)
SSE = np.sum((Y - Y_predict)**2)
print(SST)
print(SSE)
print((SST - SSE)/SST)
```
