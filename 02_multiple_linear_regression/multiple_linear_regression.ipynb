```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv=pd.read_csv(r"\Users\vaheh\aua\Fall_2024\ML\code\CS340B\02_multiple_linear_regression\Gasprice2.txt", header=None)
data = csv.to_numpy()
```

```python
Y = np.array(data[:,2])
X = np.concatenate(([np.ones(len(Y))], data[:,0:2].transpose()))
```

```python
for i in range(1, len(X)):
    plt.scatter(X[i], Y)
    plt.show()
```

```python
def get_cost(X, Y, theta):
    return np.sum((Y - theta.dot(X))**2)/(2 * len(Y))

def linear_regression(l, iter, X, Y):
    cost_history = np.array([])
    theta = np.array([0.0, 0.01, 0.02])
    for j in range(iter):
        print("-----" )
        print(theta)
        print(theta.dot(X))
        print((Y - theta.dot(X)))
        print((Y - theta.dot(X)) * X)
        theta = theta + l*np.sum((Y - theta.dot(X)) * X, axis=1)/len(Y)
        print(theta)
        cost_history = np.append(cost_history, get_cost(X, Y, theta))
    return theta, cost_history
```

```python
X_filtered = X[0:3]
print(X_filtered)
theta, cost_history = linear_regression(0.0001,3,X_filtered,Y)
Y_predict = theta.dot(X_filtered)
print("thetas result")
print(theta)
```

```python
for i in range(1, len(X)):
    plt.scatter(X[i], Y)
    plt.scatter(X[i],Y_predict)
    plt.show()
```

```python
plt.plot(cost_history)
plt.show()
```

```python
mean = np.sum(Y) / len(Y)
SST = np.sum((Y - mean)**2)
SSE = np.sum((Y - Y_predict)**2)
print(SST)
print(SSE)
print((SST - SSE)/SST)
```
