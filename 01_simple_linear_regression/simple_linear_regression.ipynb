```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y = np.array([5,17,11,8,14,5])
X = np.array([
    np.ones(len(Y)),   
    [34,108,64,88,99,51]
])
theta = np.ones(len(Y))
```

```python
plt.scatter(X[1], Y)
plt.show()
```

```python
def linear_regression(l, iter, X, Y):
    theta = np.ones(len(X))
    for j in range(iter):
        theta = theta + l*np.sum((Y - theta.dot(X)) * X, axis=1)/len(Y)
    return theta
```

```python
theta = linear_regression(0.0001,500000,X,Y)
Y_predict = theta.dot(X)
```

```python
plt.plot(X[1], Y_predict)
plt.scatter(X[1], Y)
plt.show()
```

```python
mean = np.sum(Y) / len(Y)
SST = np.sum((Y - mean)**2)
SSE = np.sum((Y - Y_predict)**2)
print(SST)
print(SSE)
print((SST - SSE)/SST)
```
